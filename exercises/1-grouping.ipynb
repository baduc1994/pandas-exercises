# Occupation

### Step 1. Import relevant libraries

```python
import pandas as pd
```

### Step 2. Check dataset from `users.txt` file

- Pay attention to what is separating the columns?
- File extension is `.txt`. This should not matter for us when importing data

### Step 3. Import data & assign it to a variable called users.

```python
users = pd.read_csv('users.txt',sep="|")
df = users.dropna()
df
```

### Step 4. Discover what is the mean age per occupation

```python
a = df.groupby(by='occupation').agg({'age': 'mean'})
a = a.round()
a
```

### Step 5.  For each occupation present the percentage of women and men

```python
df_occ_gender = df[['occupation', 'gender']]
df_occ_gender['quantity'] = 1
df_sum_each_gender = df_occ_gender.groupby(['occupation','gender']).sum()
df_sum_each_gender
```

```python
df_sum_each_gender['percentage'] = 100 * df_sum_each_gender['quantity'] / df_occ_gender.groupby('occupation')['quantity'].sum()
df_percent_gender = df_sum_each_gender
df_percent_gender.round(2)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
df_min_max_age = df.groupby(by='occupation')['age'].agg(['min','max'])
df_min_max_age
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
df_mean_age = df.groupby(['occupation','gender']).agg({'age':'mean'})
df_mean_age.round()
```

### Step 8. Discover the Male ratio per occupation and sort it from the most to the least

```python

df_male_percent = df_percent_gender.reset_index()
df_male_percent
```

```python
df_male_percent[df_male_percent['gender'].str.contains("M")].round(2).sort_values(by = 'percentage',ascending = False)
```
